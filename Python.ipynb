```python
#WAP that takes list as parameter and returns only even list
# Function to filter even numbers from a list
def get_even_list(items):
    even_list = []
    for num in items:
        if num % 2 == 0:
            even_list.append(num)
    return even_list

# Example usage
numbers = [1, 2, 3, 4, 5, 6]
result = get_even_list(numbers)
print("Even numbers:", result)
```
