```python
#string
text = "We are learning python"
#indexing: accessing elements based on the index/position.      []
print(text[0]) #indexing starts from 0
print(text[1]) #indexing starts from 1
print(text[-1]) #indexing starts from -1 last element
print(text[2])  #indexing starts from 2 space is also an element
print(text[-2]) #indexing starts from second last element
#print(text[40]) #indexing error, because there are no 41 elements
```

```python
#slicing: taking certain part/pieces of the string
text = "we are learning python"
print(text[0:4]) #0 inclusive 4 exclusive
print(text[:4]) # same as above, better
print(text[4:5])  #slicing with a gap of 1 is like indexing
print(text[5:])   #5th index to last
print(text[:-1])    #last exclusive
print(text[:])    #copy operation
print(text[::-1])   #[start:end:operation] here -1 is reverse
```
